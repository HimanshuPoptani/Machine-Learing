```python
arr_random = np.random.randint(1,100,100) # Array of 100 random integers used further 
random = arr_random.reshape(10,10) # Reshaping that array again
print(random)
Max = arr_random.max()
Min = arr_random.min()
print(f"Max value here is: {Max}")
print(f"Min value here is: {Min}")
print(f"Max value is at location: {arr_random.argmax()}") # lOCATION OF MAX VALUE
print(f"Min value is at location: {arr_random.argmin()}") # LOCATION OF MIN VALUE
```
