# DATAFRAMES
### This is the main tool for using Pandas

```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
from numpy.random import randint
```

```python
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df
```

```python
df
```

## Conditional Selections

```python
df > 0 # True at the places where element obeys the condition
```

```python
df < 0 # False where the elements dosn't obey the condition
```

```python
booldf1 = df > 0 # Just made a boolean dataframe for further use
booldf1
```

```python
df[booldf1] # This is because the elements are true
```

```python
booldf2 = df < 0 # All false values
df[booldf2]
```

### Meathod 2 (Conditional values)

```python
df[df > 0]
```

```python
df[df < 0]
```

```python
print(df['W'] > 0) # This is a series
print(df['W'] < 0) # Series
```

```python
row_A = df.loc['A']
print(f'ROW A values > 0 \n{row_A > 0}\n')
print(f'ROW A values < 0 \n{row_A < 0}\n')
```

```python
df
```

```python
# Conditional entry
df[df['W'] > 0]
```

```python
# Conditional entry
df[df['Z'] < 0] 
```

```python
result_df = df[df['Y'] > 0]
result_df # We can now also grab element out of it also
```

```python
result_df ['X']
```

```python
# Similar way of doing the same as above is:
df[df['Y'] > 0] ['X']
```

Similar way of doing the same as above ^^^^

```python
# Multiple columns for the same
df[df['Y'] > 0] [['X','Y']]
```

```python
# Similar way of doing the same as above
boolser = df['Y'] > 0
result = df[boolser]
mycols = ['X','Y']
result[mycols]
```

# Multiple conditions

```python
# AND Operator
df[(df['W'] > 0) & (df['Y'] > 0)] # Multiple conditions
```

```python
# OR Operator
df[(df['W'] > 0) | (df['Y'] > 0)] # Multiple conditions
```

# Resetting Index

```python
df
```

```python
# Indexing reset
# Not replacing the original one
# For replacing just enter inplace = "True"
df.reset_index() 
```

```python
# .split just makes it with commas
newint = 'CA NU WY OR CO'.split()
newint
```

```python
# Setting index
# For making a new column
df['new_col'] = newint
df
```

```python
# set.index
# for setting a index
# This is again not in place, set inplace = True
df.set_index('new_col')
```
