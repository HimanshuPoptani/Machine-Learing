```python
import matplotlib.pyplot as plt
```

```python
# Use to the plots in build in this notebook
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(-5,5,11)
y = x ** 2
```

```python
print(x)
print(y)
```

## Functional Meathod:

```python
# Functional 
plt.plot(x,y) # -g is just for a green color
plt.show() # Only if not using Jupyter notebook
```

```python
plt.plot(y,x)
```

```python
# lABELING
plt.plot(x,y)
plt.xlabel('Some integers') # X Label
plt.ylabel('Square') # Y Label
plt.title("Square Function") # Title
```

```python
# Multiplots
plt.subplot(1,2,1) # subplot(no. of rows, no.of cols, plot no. refering to)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'g-')
```

## Object Oriented Meathod:

```python
# OO
fig = plt.figure() # This is a empty canvas for making

# add_axes([left_axes, bottom_axes, width, hieght])
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
# Labelling
# It is different from Functional Meathod
axes.set_title("Squre function")
axes.set_xlabel("Integers")
axes.set_ylabel("Square")
```

```python
# Multiplotting
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y) # Function 1
axes2.plot(y,x) # Function 2

axes1.set_ylabel("Square Function")
axes2.set_xlabel("Square Function")

axes1.set_title("LARGER PLOT")
axes2.set_title("SMALLER PLOT")
```
